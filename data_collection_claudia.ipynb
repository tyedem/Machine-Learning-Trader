```python
# Import all libraries and dependencies

import datetime
import pandas as pd
import yfinance as yf
```

```python
#Define the tickers to download data of the stocks.
tickers = ["AAPL", "LAC.TO", "CIX.TO", "SJR-B.TO", "L.TO", "TOY.TO"]

# Fetch tickers data from 1/1/2012 until 3/3/2022.
start_date = datetime.date(2012,1,1)
end_date = datetime.date(2022,3,3)
interval = '1d'
period = round((end_date - start_date).days/365,0)
if interval == "1d":
    annual_interval = 252
```

```python
# Use yfinance to download the data
yf.Tickers(tickers)
stock_data = yf.download(tickers, start = start_date, end = end_date, interval = interval)

# Check the data type
type(stock_data)
```

```python
# Review data
stock_data
```

```python
# Select closing prices only
price = stock_data[["Close"]]
price.head()
```
